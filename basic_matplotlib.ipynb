```python
def fun(t):
    return (np.exp(-t)*np.cos(2*np.pi*t))
```

```python
t1=np.arange(0,5,0.1)
t2=np.arange(0,5,0.3)
plt.figure(1)
plt.subplot(221)
plt.plot(t1,fun(t1),color='b')
plt.grid()
plt.subplot(222)
plt.plot(t2,fun(t2))
plt.subplot(223)
plt.plot(t2,fun(t2))
plt.subplot(224)
plt.plot(t2,fun(t2))

plt.figure(2)
plt.subplot(221)
plt.plot(t1,fun(t1),color='r')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import numpy as random
data1=np.random.rand(5)*10
data2=np.random.rand(5)*10
data3=np.random.rand(5)*10
locs=np.arange(1,len(data1)+1)
width=0.27
plt.bar(locs,data1)
plt.bar(locs+width,data2,color='green',width=width)
plt.bar(locs+width*2,data3,width=width,color='red')
plt.xticks(locs+width*1.5,locs)
plt.show()
```
